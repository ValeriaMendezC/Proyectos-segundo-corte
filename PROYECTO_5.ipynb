#5 Divida el dataset en dos grupos a partir de la variable Attrition Flag. Verifique si hay diferencias en la media o en la mediana significativas en cuanto a Customer Age y en cuanto a Dependent Count para los dos grupos.

```python
data3 = data[['Attrition_Flag','Customer_Age', 'Dependent_count']]
data3.replace({'Attrition_Flag':{'Existing Customer':1, 'Attrited Customer' :0}}, inplace = True)
Existing_Customer = []
Attrited_Customer = []
for i in  data3['Attrition_Flag']:
  Existing_Customer.append(i)
  if i == i:
    if i == 1:
      i=0
      Attrited_Customer.append(i)
    else:
      i=1
      Attrited_Customer.append(i)
      

data3['Existing_Customer'] = Existing_Customer
data3['Attrited_Customer'] = Attrited_Customer
```

```python
#data3
```

```python
data3.groupby('Existing_Customer').agg({'Customer_Age':['mean','median'], 'Dependent_count':['mean','median']})
```

```python
data3.groupby('Attrited_Customer').agg({'Customer_Age':['mean','median'], 'Dependent_count':['mean','median']})
```

Prueba de hipótesis

H0 = No hay diferencia entre las medias

HA = Si hay diferencias entre las medias

```python
mean_E_C = stats.ttest_ind(data3.groupby('Existing_Customer').agg({'Customer_Age':['mean']}), data3.groupby('Existing_Customer').agg({'Dependent_count':['mean']}))
mean_E_C #Es la media de Exiting
```

```python
if mean_E_C[1] > 0.05: #Es la media de Exiting
  print('se acepta la hipótesis nula, por lo que no existe diferencia entre las medias')
else:
  print('Se rechaza la hipótesis nula, por lo que si existe diferencia entre las medias')
```

**Conclusión intermedia:** a partir del p valor 2.08859191e-05<0.05 se rechaza la hipótesis nula, por lo que exite diferencia de medias entre Customer_Age y Dependent_Count para los datos existentes dentro de Existing_Customer

```python
mean_A_C = stats.ttest_ind(data3.groupby('Attrited_Customer').agg({'Customer_Age':['mean']}), data3.groupby('Attrited_Customer').agg({'Dependent_count':['mean']}))
mean_A_C #Es la media de Attrited
```

```python
if mean_A_C[1] > 0.05: #Es la media de Attrited
  print('se acepta la hipótesis nula, por lo que no existe diferencia entre las medias')
else:
  print('Se rechaza la hipótesis nula, por lo que si existe diferencia entre las medias')
```

**Conclusión intermedia:** a partir del p valor 2.08859191e-05<0.05 se rechaza la hipótesis nula, por lo que exite diferencia de medias entre Customer_Age y Dependent_Count para los datos existentes dentro de Attrited_Customer

```python
median_E_C = stats.ttest_ind(data3.groupby('Existing_Customer').agg({'Customer_Age':['median']}), data3.groupby('Existing_Customer').agg({'Dependent_count':['median']}))
median_E_C #Es la mediana de Exiting
```

```python
if median_E_C[1] > 0.05:#Es la mediana de Exiting
  print('se acepta la hipótesis nula, por lo que no existe diferencia entre las medianas')
else:
  print('Se rechaza la hipótesis nula, por lo que si existe diferencia entre las medianas')
```

**Conclusión intermedia:** a partir del p valor 0.00012622<0.05 se rechaza la hipótesis nula, por lo que exite diferencia de medianas entre Customer_Age y Dependent_Count para los datos existentes dentro de Existing_Customer

```python
median_A_C = stats.ttest_ind(data3.groupby('Attrited_Customer').agg({'Customer_Age':['median']}), data3.groupby('Attrited_Customer').agg({'Dependent_count':['median']}))
median_A_C #Es la mediana de Attrited
```

```python
if median_A_C[1] > 0.05:#Es la mediana de Attrited
  print('se acepta la hipótesis nula, por lo que no existe diferencia entre las medianas')
else:
  print(f'Se rechaza la hipótesis nula, por lo que si existe diferencia entre las medianas')
```

**Conclusión intermedia:** a partir del p valor 0.00012622<0.05 se rechaza la hipótesis nula, por lo que exite diferencia de medianas entre Customer_Age y Dependent_Count para los datos existentes dentro de Attrited_Customer
